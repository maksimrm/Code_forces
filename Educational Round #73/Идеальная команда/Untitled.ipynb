```python
q = 10**2
```

```python
ans = []
for i in range(q):
    for j in range(q):
        for k in range(q):
            ans.append([i, j, k])
```

```python
len(ans)
```

```python
print(ans[:10])
```

```python
def print_num(ans, start, end):
    print(-start + end)
    for i in range(start, end):
        print("{} {} {}".format(ans[i][0], ans[i][1], ans[i][2]))
```

```python
print_num(ans, 10000, 20000)
```
